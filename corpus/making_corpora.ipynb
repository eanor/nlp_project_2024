```python
def lemmatizator(corpus):
  result=''
  doc=Doc(corpus)
  doc.segment(segmenter)
  doc.tag_morph(morph_tagger)
  for token in doc.tokens:
     token.lemmatize(morph_vocab)
  for token in doc.tokens:
    if token.pos!='PUNCT':
      result=result+' '+token.lemma
    else:
      result=result+token.lemma
  return(result[1:])
```

```python
def del_punct(s):
  new_s = re.sub(r'[^\w\s]', '', s)
  return new_s
```

```python
def making_corpus(input_texts, output_texts):
  df = pd.read_csv (input_texts, sep="\t")
  corpus=[]
  for sent in df['sent1']:
    sent1=re.sub("<b><i>|</i></b>","",sent)
    sent1_lemm=lemmatizator(del_punct(sent1))
    print(sent1_lemm)
    corpus.append(sent1_lemm)
  for sent2 in df['sent2']:
    sent3=re.sub("<b><i>|</i></b>","",sent2)
    sent3_lemm=lemmatizator(del_punct(sent3))
    corpus.append(sent3_lemm)
  with open(output_texts, "w") as file:
    for  line in corpus:
        file.write(line + '\n')
```
