```python
def collect_sentences(words, sents_fp):

    with open(sents_fp, "r") as fh:
        lines = fh.read().split("\n")

    word_sents = {word: [] for word in words}

    for sent in tqdm(lines, desc="Собираем предложения"):
        for word in words:
            if word in sent:
                word_sents[word].append(sent)

    return word_sents
```

```python
def find_first_seq_ics(arr, seq):
    seq_pos = 0
    current_start = 0

    for ix, elem in enumerate(arr):
        if elem == seq[seq_pos]:
            seq_pos += 1
            if seq_pos == len(seq):
                return np.arange(current_start, current_start + seq_pos)
        else:
            current_start = ix + 1
            seq_pos = 0

    return []
```

```python
def extract_representations(c1_path, c2_path, c3_path, targets_path, experiment_dir, device="cpu", **kwargs):
    """ Extracts last hidden layer values for all target words in a datasets."""

    rep_dir_c1 = experiment_dir + "word_representations/c1/"
    rep_dir_c2 = experiment_dir + "word_representations/c2/"
    rep_dir_c3 = experiment_dir + "word_representations/c3/"

    os.makedirs(rep_dir_c1, exist_ok=True)
    os.makedirs(rep_dir_c2, exist_ok=True)
    os.makedirs(rep_dir_c3, exist_ok=True)

    with open(targets_path, "r") as fh:
        targets = fh.read().splitlines()

    sents_c1 = collect_sentences(targets, c1_path)
    sents_c2 = collect_sentences(targets, c2_path)
    sents_c3 = collect_sentences(targets, c3_path)

    bert_dir = experiment_dir + "bert/"
    device = torch.device(device)

    tokenizer, model = load_local_bert(bert_dir, device, output_hidden_states=True)
    model.eval()

    save_representations(sents_c1, model, tokenizer, device, rep_dir_c1)
    save_representations(sents_c2, model, tokenizer, device, rep_dir_c2)
    save_representations(sents_c3, model, tokenizer, device, rep_dir_c3)


def save_representations(word_sents, model, tokenizer, device, rep_dir, max_sql=512):

    for word, sents in tqdm(word_sents.items()):
        
        word_hidden_states = []
        word_tokens = tokenizer.encode(word)[1:-1]

        for sent in sents:
            
            hidden_states, encoded = get_hidden_from_sent(sent, model, tokenizer, device, max_sql)
            
            if len(word_tokens) > 0:
                word_token_ics = find_first_seq_ics(encoded, word_tokens)
            else:
                word_token_ics = [np.argmax(encoded == word_tokens[0])]

            if len(word_token_ics) > 0:
                mean_state_last_layer = hidden_states[-1, word_token_ics, :].mean(axis=0)
                word_hidden_states.append(mean_state_last_layer)

        np.save(rep_dir + word + ".npy", np.array(word_hidden_states))


def get_hidden_from_sent(sent, model, tokenizer, device, max_sql):

    tokenized = [tokenizer.cls_token] + tokenizer.tokenize(sent)[:(max_sql - 2)] + [tokenizer.sep_token]
    encoded = tokenizer.convert_tokens_to_ids(tokenized)
    padded = np.array(encoded + [tokenizer.pad_token_id] * (max_sql - len(encoded))).reshape(1, -1)

    token_type_ids = np.zeros_like(padded)
    attention_mask = (~(padded == tokenizer.pad_token_id)).astype(int)
    labels = np.array([1]).astype(int)

    inputs = {
        "input_ids": torch.from_numpy(padded).to(device),
        "attention_mask": torch.from_numpy(attention_mask).to(device),
        "token_type_ids": torch.from_numpy(token_type_ids).to(device),
        "labels": torch.from_numpy(labels).to(device)
    }

    hidden_matrix = np.vstack([v.cpu().detach().numpy() for v in model(**inputs)[-1]])

    return hidden_matrix[:, :len(encoded), :], np.array(encoded)
```

```python
def load_rep_dict(rep_dir, target):

    rep_dict = {}
    rep_dict[target] = np.load(rep_dir + target + ".npy")
    assert rep_dict[target].size > 0, "No representations saved for word '{}' - check both corpora for occurrences!".format(target)

    return rep_dict
```

```python
from scipy.spatial import distance

def get_cosine_dist(Xs, Ys):
    Xs = np.mean(Xs, axis=0)
    Ys = np.mean(Ys, axis=0)
    return distance.cosine(Xs, Ys)
```

```python
def compare_context_dependent_representations(targets_path, experiment_dir):

    with open(targets_path, "r") as fh:
        targets = fh.read().splitlines()
        
    dists_1_2 = []
    dists_2_3 = []
    dists_1_3 = []
    
    for target in tqdm(targets):
        
        c1_reps = load_rep_dict(experiment_dir + "word_representations/c1/", target)
        c2_reps = load_rep_dict(experiment_dir + "word_representations/c2/", target)
        c3_reps = load_rep_dict(experiment_dir + "word_representations/c3/", target)
        
        dist_1_2 = get_cosine_dist(c1_reps[target], c2_reps[target])
        dist_2_3 = get_cosine_dist(c2_reps[target], c3_reps[target])
        dist_1_3 = get_cosine_dist(c1_reps[target], c3_reps[target])

        dists_1_2.append(dist_1_2)
        dists_2_3.append(dist_2_3)
        dists_1_3.append(dist_1_3)

    pd.DataFrame({"word": targets, "change_1_2": dists_1_2, "change_2_3": dists_2_3, "change_1_3": dists_1_3}).to_csv(experiment_dir + "prediction.tsv", sep="\t", index=False, header=False)
```

```python
compare_context_dependent_representations(targets_path, experiment_dir)
print("Результаты сравнения репрезетаций слов для каждой пары эпох получены!")
```
